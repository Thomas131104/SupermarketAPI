# Step 1: Đưa từng nhóm mặt hàng lên

```python
categories = [
    {"name": "Đồ gia dụng", "discount": 10},   # 10% giảm giá
    {"name": "Đồ dùng nhà bếp", "discount": 5},
    {"name": "Văn phòng phẩm", "discount": 0}
]
```

```python
import requests

BASE_URL = "https://5765e62b-7a02-4a62-8ec9-912dd061f20d-00-3gsihqdsrqdxh.sisko.replit.dev"

category_ids = []

for cat in categories:
    res = requests.post(f"{BASE_URL}/categories/", json=cat)
    print(res.status_code, res.text)  # <-- check response
    data = res.json()
    if "id" in data:
        category_ids.append(data["id"])
    else:
        print("No 'id' in response:", data)

print("Category IDs:", category_ids)
```

# Step 3: Kiểm tra

```python
# Lấy tất cả Category

response = requests.get(f"{BASE_URL}/categories/")
print(response.status_code)
print(response.json())
```

```python
# Lấy category theo ID

id = 1
response = requests.get(f"{BASE_URL}/categories/{id}")
print(response.status_code)
print(response.json())
```

```python
# Giảm giá

update_data = {"discount": 15}  # giảm giá mới
response = requests.put(f"{BASE_URL}/categories/{id}", json=update_data)
print(response.status_code)
print(response.json())
```

```python
# Xóa category
response = requests.post(f"{BASE_URL}/categories/", json = {"name" : "Máy vi tính"})

response = requests.delete(f"{BASE_URL}/categories/4")
print(response.status_code)
print(response.json())  # {'detail': 'Category deleted'}
```
